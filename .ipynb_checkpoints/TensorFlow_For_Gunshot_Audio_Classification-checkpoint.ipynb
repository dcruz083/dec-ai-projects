# Task 1: PubMed 20k RCT: Dataset for Sequential Sentence Classification

```python
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Bidirectional, SpatialDropout1D 
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dropout
import numpy as np
from sklearn.metrics import classification_report
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GlobalMaxPooling1D
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Reshape
```

```python
def readFile(filename): #function to read files
    return pd.read_csv(filename)
```

```python
#Loading datasets
train_df = readFile('train.csv') 
val_df = readFile('val.csv')
test_df = readFile('test.csv')
train_df
```

```python
#showing sizes for datasets
print(train_df.shape, val_df.shape,test_df.shape)

#unique labels in train set
print(train_df.iloc[:,-1].unique())
```

```python
col_name='abstract_text'

#function for preprocessing text data
def cleanData(df, col_name):
    df['cleaned_text'] = df[col_name].str.lower()
    return df

#cleaning data
train_df = cleanData(train_df,col_name)
val_df = cleanData(val_df,col_name)
test_df = cleanData(test_df,col_name)

#printing train df with clean text column
train_df
```

```python
# Tokenization and padding parameters
vocab_size= 60000  # Maximum number of words to keep
maxlen = 64  # Maximum sequence length
trunc_type = 'post'
padding_type = 'post'
ovv_tok = "<OOV>"
embedding_dim = 128  # Dimension of word embeddings
```

```python
#Tokenizing text
tokenizer = Tokenizer(num_words=vocab_size, oov_token=ovv_tok)
tokenizer.fit_on_texts(train_df['cleaned_text'])
word_idx = tokenizer.word_index
```

```python
#function to get sequence padding
def text_padder(df, col_name, maxlen, padding_type, trunc_type):
    sequences = tokenizer.texts_to_sequences(df[col_name])
    padded_sequences = pad_sequences(sequences, 
                                     maxlen=maxlen, 
                                     padding=padding_type, 
                                     truncating=trunc_type)
    return padded_sequences

#function to do label encoding
def label_enconder(df, col_name):
    labels = pd.get_dummies(df[col_name]).values
    return labels
```

```python
#Setting column names for sequence padding and label encoder
col_name_padder = 'cleaned_text'
col_name_label = 'target'

#getting padded sequences and dummy labels
train_padded = text_padder(train_df,col_name_padder, maxlen, padding_type, trunc_type)
train_labels = label_enconder(train_df,col_name_label)

val_padded = text_padder(val_df,col_name_padder, maxlen, padding_type, trunc_type)
val_labels = label_enconder(val_df,col_name_label)

test_padded = text_padder(test_df,col_name_padder, maxlen, padding_type, trunc_type)
test_labels = label_enconder(test_df,col_name_label)
```

```python
# Define the sequential sentence classification model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=maxlen))
model.add(Bidirectional(LSTM(128, return_sequences=True)))  # Use return_sequences=True for stacking LSTM layers

# First dense layer
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # Add dropout for regularization

# Second LSTM layer
model.add(Bidirectional(LSTM(64)))

# Additional dense layers
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))  # Add dropout for regularization
model.add(Dense(32, activation='relu'))
model.add(Dense(len(train_df['target'].unique()), activation='softmax'))
```

```python
# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# Train the model with separate validation data
epochs = 5
model.fit(train_padded, train_labels, epochs=epochs, validation_data=(val_padded, val_labels))
```

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(test_padded, test_labels)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
def model_evaluation(model, test_padded, y_true, class_labels):
    # Getting and Converting predictions to class labels
    y_pred = np.argmax(model.predict(test_padded), axis=1)
    
    print(classification_report(y_true, y_pred, target_names=class_labels))

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
train_encoded_labels = label_encoder.fit_transform(train_df['target'])
test_encoded_labels = label_encoder.transform(test_df['target'])
    
print('\nModel Evaluation:')
model_evaluation(model, test_padded, test_encoded_labels, sorted(test_df['target'].unique()))
```

#### The sequential classification model for classification of sentences in the context of their position in abstracts of biomedical research papers achieved a 80% accuracy when increasing the filter for bidirectional LSTM layers and reducing the number of epochs for fitting.

```python
import os
import numpy as np
import librosa
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Reshape
from tensorflow.keras.utils import to_categorical
```

```python
# Function to read sounds and convert to spectrograms
def read_data(folder_path):
    labels = []
    spectrograms = []

    for label in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, label)
        if os.path.isdir(subfolder_path):
            for file in os.listdir(subfolder_path):
                file_path = os.path.join(subfolder_path, file)
                if file_path.endswith('.wav'):
                   # print(file_path)
                    y, sr = librosa.load(file_path)
                    S = librosa.stft(y)
                    S_mag = np.abs(S)
                    S_dB = librosa.amplitude_to_db(S_mag, ref=np.max)
                    spectrograms.append(S_dB)
                    labels.append(label)
    
    return spectrograms, labels
# Function to pad or trim a 2D array to a desired shape
def pad2d(a, desired_size):
    rows, cols = a.shape
    padded_a = np.zeros((desired_size, desired_size))
    rows_to_copy = min(rows, desired_size)
    cols_to_copy = min(cols, desired_size)
    padded_a[:rows_to_copy, :cols_to_copy] = a[:rows_to_copy, :cols_to_copy]
    return padded_a
```

```python
# Read spectrograms and labels
spectrograms, labels = read_data(main_directory)
```

```python
# BASELINE: Create CNN model
def create_cnn_model(input_shape, num_classes):
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())
    model.add(Flatten())
    
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))

    return model
```

```python
# Preprocess data
desired_spectrogram_size = 128
spectrograms = np.array([pad2d(s, desired_spectrogram_size) for s in spectrograms])
spectrograms = np.expand_dims(spectrograms, axis=-1)  # Add channel dimension
print(spectrograms.shape)
label_dict = {label: i for i, label in enumerate(set(labels))}
y = np.array([label_dict[label] for label in labels])
y = to_categorical(y)  # One-hot encoding
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(spectrograms, y, test_size=0.2, random_state=42)
```

### Baseline Model

```python
# Define input shape and number of classes
input_shape = X_train[0].shape
num_classes = y.shape[1]

# Create and compile the model
baseline_model = create_cnn_model(input_shape, num_classes)
baseline_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
baseline_model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))

# Save the model
```

```python
# Evaluate the model on the test set
loss, accuracy = baseline_model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
y_pred_baseline= np.argmax(baseline_model.predict(X_test), axis=1)

# Convert label probabilities to class labels
y_test_classes = np.argmax(y_test, axis=1)

print(y_test_classes.shape)
print(y_pred_baseline.shape)
```

```python
print(classification_report(y_test_classes, y_pred_baseline, target_names=np.unique(labels)))
```

### First custom model CNN Model

```python
#first custom model reducing the number of Conv2D filters
def create_custom_model(input_shape, num_classes):
    model = Sequential()
    model.add(Conv2D(8, (3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())
    model.add(Conv2D(16, (3, 3), activation='relu'))  # Convolutional layer with 16 filters, kernel size 3x3
    model.add(MaxPooling2D((2, 2)))  # Max pooling layer with pool size 2x2
    model.add(BatchNormalization())
    model.add(Conv2D(32, (3, 3), activation='relu'))  # Convolutional layer with 32 filters, kernel size 3x3
    model.add(MaxPooling2D((2, 2)))  # Max pooling layer with pool size 2x2
    model.add(BatchNormalization())
    model.add(Flatten())
    
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(16, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    
    return model
```

```python
# Define input shape and number of classes
input_shape = X_train[0].shape
num_classes = y.shape[1]
#num_classes
# Create and compile the model
custom_model = create_custom_model(input_shape, num_classes)
custom_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
custom_model.summary()
custom_model.fit(X_train, y_train, epochs=300, validation_data=(X_test, y_test))
```

```python
# Evaluate the model on the test set
loss, accuracy = custom_model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

### Second custom CNN model

```python
#second custom CNN model increasing the Conv2D filters and removing a dense layer
def create_custom_model(input_shape, num_classes):
    model = Sequential()
    model.add(Conv2D(64, (3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())

    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())

    model.add(Conv2D(256, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())

    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))

    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))

    model.add(Dense(num_classes, activation='softmax'))
    return model
```

```python
# Define input shape and number of classes
input_shape = X_train[0].shape
num_classes = y.shape[1]
#num_classes
# Create and compile the model
custom_model_2 = create_custom_model(input_shape, num_classes)
custom_model_2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
custom_model_2.summary()
custom_model_2.fit(X_train, y_train, epochs=200, validation_data=(X_test, y_test))
```

```python
# Evaluate the model on the test set
loss, accuracy = custom_model_2.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
y_pred_baseline= np.argmax(custom_model_2.predict(X_test), axis=1)

# Convert label probabilities to class labels
y_test_classes = np.argmax(y_test, axis=1)
print(classification_report(y_test_classes, y_pred_baseline, target_names=np.unique(labels)))
```

### Third custom model using CNN and LSTM layers

```python
#third custom model using CNN and LSTM layers
def create_custom_model(input_shape, num_classes):
    model = Sequential()
    model.add(Conv2D(64, (3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())
    
    # Adding an LSTM layer
    model.add(Flatten())
    model.add(Dense(64, activation='relu'))
    model.add(Reshape((1, 64)))
    model.add(Bidirectional(LSTM(32, return_sequences=False)))

    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.5))  # Regularization

    model.add(Dense(num_classes, activation='softmax'))
    return model
```

```python
# Define input shape and number of classes
input_shape = X_train[0].shape
num_classes = y.shape[1]
#num_classes
# Create and compile the model
custom_model_3 = create_custom_model(input_shape, num_classes)
custom_model_3.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
custom_model_3.summary()
custom_model_3.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))
```

```python
# Evaluate the model on the test set
loss, accuracy = custom_model_3.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
y_pred_baseline= np.argmax(custom_model_3.predict(X_test), axis=1)

# Convert label probabilities to class labels
y_test_classes = np.argmax(y_test, axis=1)
print(classification_report(y_test_classes, y_pred_baseline, target_names=np.unique(labels)))
```

### Fourth custom model using CNN and LSTM layers

```python
#4th custom model using CNN and LSTM layers
def create_custom_model(input_shape, num_classes):
    model = Sequential()
    model.add(Conv2D(64, (3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())
    
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(BatchNormalization())

    # Add an LSTM layer
    model.add(Flatten())
    model.add(Dropout(0.5))  # Regularization

    model.add(Dense(64, activation='relu'))
    model.add(Reshape((1, 64)))
    model.add(Bidirectional(LSTM(32, return_sequences=False)))

    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.5))  # Regularization

    model.add(Dense(num_classes, activation='softmax'))
    return model
```

```python
# Define input shape and number of classes
input_shape = X_train[0].shape
num_classes = y.shape[1]
#num_classes
# Create and compile the model
custom_model_4 = create_custom_model(input_shape, num_classes)
custom_model_4.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
custom_model_4.summary()
custom_model_4.fit(X_train, y_train, epochs=200, validation_data=(X_test, y_test))
```

```python
# Evaluate the model on the test set
loss, accuracy = custom_model_4.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```

```python
y_pred_baseline= np.argmax(custom_model_4.predict(X_test), axis=1)

# Convert label probabilities to class labels
y_test_classes = np.argmax(y_test, axis=1)
print(classification_report(y_test_classes, y_pred_baseline, target_names=np.unique(labels)))
```

## Report:
### Research: 
#### The research involved developing a deep learning neural network capable of classifying gunshot audio based on the type of firearm and its orientation. 

### Approach:
#### The approach used a convolutional neural network architecture that features Conv2D layers. These layers helped to extract hierarchical features from two-dimensional spectrogram data and allowed spatial analysis across the frequency and time dimensions. This analysis enabled the network to understand high-level representations of audio features. After the convolutional layers, a series of densely connected layers processed the learned features further. This helped the model to capture intricate patterns and relationships between the extracted features. Ultimately, this facilitated classification of the audio into distinct firearm types. 

### Experiments and results: 
#### The initial custom model was adjusted to decrease the Conv2D filters and an addition of 2 dense layers with dropout of 0.5 after each dense layer output to prevent overfitting. After compiling the model with 300 epochs, the overall accuracy decreased by 6% from the baseline_model. However it saw a decrease in loss by 0.0291 against the test dataset compared to the baseline_model. 
#### By trial and error, the filter sizes are adjusted along with the removal of a dense layer to build custom_model_2. Trained with 200 epochs, it yields similar results in terms of accuracy as the baseline_model however it scores a much lower loss score compare to baseline_model.
#### After combining Conv2D and LSTM layers into a single custom_model_3 and training it for 100 epochs, there was 12% decrease in accuracy compared to the custom_model_2 which only uses convolutional layers. To improve the performance of the combined CNN and LSTM architecture, I constructed a fourth model called custom_model_4 that includes an additional 128 filter Conv2D layer and a 64 neuron dense layer. This model significantly increased in performance and reached an accuracy similar to that of custom_model_2 and baseline_model, which is 91%. Although both models have different training methods, they perform similarly. 
#### To choose the best model for the given dataset, we can leverage the loss score between custom_model_2 and custom_model_4 which are 0.5575 and 0.8063 respectively. It is also important to note that custom_model_2 performs better than custom_model_4 despite having the same accuracy. Both models performed similarly in accuracy when compared to the baseline_model, but showed better performance in loss. Additionally, training time is a factor to consider, as both models were trained for 200 epochs, custom_model_2 takes around 20 seconds/epoch, which is 4 seconds/epoch slower than custom_model_4.

### Interesting findings/challenges:
#### An intriguing trend emerged as the different experiments with increasing filter sizes in the convolutional layers of our models were trained. Notably, as the filter sizes grew, the training time for the model proportionally increased. Larger filter sizes in convolutional layers of our model increase computational complexity and training time. However, higher epochs improve accuracy, as the model adapts to the increased complexity. This trade-off between computational cost and model performance should be considered in practical deployment scenarios.

#### In this effort, neural networks were trained trained to classify gunshot audio based on the type of firearms. Different models were experimented with by combining Conv2D and LSTM layers, and the number of filters and neurons in the layers was varied. The experiments demonstrated the effectiveness of using deep learning neural networks in classifying gunshot audio based on the type of firearm. The research also highlighted the importance of considering computational cost and model performance trade-offs when deploying such models in practical scenarios. Overall, this work demonstrates the potential of neural networks for classifying gunshot audio based on firearm type and explore application of AI in the field of gunshot detection and analysis.
