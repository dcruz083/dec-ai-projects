# Intializing Testing and Evaluation functions to be used in the implementations of PCA and KPCA throughout the project

```python
# Function to calculate distance between two points
def dis(x1, x2):
    return np.linalg.norm(x1 - x2)

# Function to perform classification 
def myclassifier(Train, Trainlabel, Test):
    pred = []

    for testpoint in Test:
        pred_dis = []
        for trainpoint in Train:
            pred_dis.append(dis(testpoint, trainpoint))

        pred.append(np.argmin(pred_dis))  

    return np.array(pred)
```

# Below function is to calculte the accuracy

```python
def calculate_accuracy(true_labels, predicted_labels):
    # Ensure that the true labels and predicted labels have the same length
    if len(true_labels) != len(predicted_labels):
        raise ValueError("Length of true_labels and predicted_labels must be the same.")

    # Count the number of correct predictions
    correct_predictions = sum(1 for true, predicted in zip(true_labels, predicted_labels) if true == predicted)

    # Calculate accuracy as the ratio of correct predictions to total predictions
    accuracy = correct_predictions / len(true_labels)

    return accuracy
```

# Part 1: Principal Component Analysis (PCA)

# 1.1 Implement PCA from Scratch  
# Create PCA function that takes two inputs: dataset, and k= number of principle component, and return the transform data

```python
#1.1 a
def my_pca(Data,k):
    
    #getting mean of training data
    mean = np.mean(Data, axis=0)
    
    #centering data
    centeredData = Data - mean
    
    #calculating the convariance matrix
    cov_matrix = np.cov(centeredData, rowvar=False)
    
    #Eigendecomposition on the Covariance Matrix
    eigenvalues, eigenvector = np.linalg.eig(cov_matrix)
    
    #Sort the Principal Components
    sorted_idx = np.argsort(eigenvalues)[::-1]

    # utilizing the sort order to sort eigenvalues and eigenvectors
    eigenvalues = eigenvalues[sorted_idx]
    eigenvector = eigenvector[:, sorted_idx]
    
    #1.1 c
    #Calculating the Explained Variance
    explained_variance = eigenvalues / np.sum(eigenvalues)
    
    # Calculate the cumulative explained variance
    total_explained_variance = sum(explained_variance[:k])
    print(f"Total Explained Variance with k=",k, " components:", total_explained_variance)#printing total variance to check for a min of 95% coverage
    
    #Projecting the data onto the selected components
    data_my_pca = np.dot(centeredData, eigenvector[:,:k])
    
    print("PCA with NumPy:")
    print(data_my_pca)
    return data_my_pca, eigenvector[:,:k], mean
```

# 1.1 b - Apply my_pca 

```python
k=110 #establishing the number of principal components to be used in the implementation of pca

#applying my_pca to get projected data, eigen vector and the training data mean
mypca_training, eigenvector, meanTrain= my_pca(train_features, k=k)

#3.1 a/b
#using centered testing data and eigen vector to transform the dataset 
mypca_testing = np.dot(test_features - meanTrain,eigenvector)

#3.2 a
#using the prediction classifier to get predicted outcomes 
predicted_mypca = train_label[myclassifier(mypca_training,train_label,mypca_testing)]

#3.2 b
#using the accuracy function to get pca accuracy results on predicted_mypca 
mypca_acc = calculate_accuracy(test_label, predicted_mypca)

print(f'Accuracy result of my pca implementation with {k} prinpical components: {mypca_acc}')
```

# Part 2: Kernel PCA (KPCA)

# Create kernel PCA function that takes two inputs: dataset, and k= number of principle component, and return the transform data. In addition, you need to create three other function one for rbf_kernel, one for polynomial_kernel, and one for linear_kernel

```python
#defining rbf_kernel
def rbf_kernel(x, y, gamma):
    return np.exp(-gamma * np.linalg.norm(x - y)**2)
#defining poly_kernel
def poly_kernel(x, y, degree):
    return (np.dot(x, y) + 1)**degree
#defining linear_kernel
def linear_kernel(x, y):
    return np.dot(x, y)

def my_kpca(data, n_components, kernel_type, kernel_param):
    
    n_samples = data.shape[0]
    K = np.zeros((n_samples, n_samples))
    #if statement to check for kernel_type parameter
    for i in range(len(data)):
        for j in range(len(data)):
            if kernel_type == 'rbf':
                K[i, j] = rbf_kernel(data[i], data[j], gamma=kernel_param)
            elif kernel_type == 'poly':
                K[i, j] = poly_kernel(data[i], data[j], degree=kernel_param)
            elif kernel_type == 'linear':
                K[i, j] = linear_kernel(data[i], data[j])
            else:
                raise ValueError("Invalid kernel type. Use 'rbf', 'poly', or 'linear'.")
    # Center the kernel matrix
    one_n = np.ones((n_samples, n_samples)) / n_samples
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)
    #print(K.shape)
    # Compute eigenvectors and eigenvalues
    eigvals, eigvecs = np.linalg.eigh(K)
    
    # Get the top n_components eigenvectors
    # Sort eigenvalues and eigenvectors in descending order
    sorted_idx = np.argsort(eigvals)[::-1]
    eigvals = eigvals[sorted_idx]
    eigvecs = eigvecs[:, sorted_idx]
    #print(eigvals)
    #print(eigenvalues)
    
    # Apply the transformation to the data
    transformed_data = np.dot(K, eigvecs[:,:n_components])
    
    return transformed_data, eigvecs[:,:n_components]

#function to transform test data with eigvecs obtained from transforming the training data
def transform_with_kpca(data, kernel_type, kernel_param, eigvecs):
    n_samples = data.shape[0]
    K = np.zeros((n_samples, len(eigvecs)))

    #if statement to check for kernel_type parameter
    for i in range(n_samples):
        for j in range(len(eigvecs)):
            if kernel_type == 'rbf':
                K[i, j] = rbf_kernel(data[i], train_features[j], gamma=kernel_param)
            elif kernel_type == 'poly':
                K[i, j] = poly_kernel(data[i], train_features[j], degree=kernel_param)
            elif kernel_type == 'linear':
                K[i, j] = linear_kernel(data[i], data[j])
            else:
                raise ValueError("Invalid kernel type. Use 'rbf', 'poly', or 'linear'.")

    # Center the kernel matrix
    one_n = np.ones((n_samples, len(eigvecs))) / len(eigvecs)
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)

    # Apply the transformation using the previously computed eigenvectors
    transformed_data = np.dot(K, eigvecs)

    return transformed_data
```

## It is a common misconception in machine learning that more features always lead to better accuracy or performance. In reality, adding more features can sometimes result in the opposite effect. This is known as the curse of dimensionality. To combat it, there are dimensionality reduction techniques such as PCA (Principal Component Analysis) and KPCA (Kernel Principal Component Analysis). These techniques aim to transform high-dimensional data into a lower-dimensional space while preserving important information. PCA does this by finding principal components (orthogonal axes) that capture the maximum variance in the data. KPCA extends on PCA by using kernel functions to perform non-linear transformations, which can capture more complex relationship data that PCA cannot. Both techniques improve classification tasks by reducing the noise and shifting the focus to the most imformative features. KPCA takes a step further mapping the data into a higher-dimensional space in non-linear classifications tasks.
